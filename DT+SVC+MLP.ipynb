```python
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
from classifier_wrapper import ClassifierWrapper


DT = DecisionTreeClassifier()
SVCGL = SVC(kernel="rbf", C=8, probability=True, break_ties=True) # best result
MLP = MLPClassifier(activation="logistic", solver="sgd", learning_rate= "constant", learning_rate_init= 0.01, max_iter = 500, shuffle= False) # best result
DT_SCVGL_MLP =  ClassifierWrapper(VotingClassifier(
    estimators=[('DT', DT), ('SVC', SVCGL), ('MLP', MLP)],
    voting='soft'))
```

```python
params  = DT_SCVGL_MLP.calc_params()
params
```
