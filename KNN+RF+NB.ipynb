```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from classifier_group import ClassifierGroup
from sklearn.naive_bayes import GaussianNB
```

```python
from classifier_wrapper import ClassifierWrapper


KNN = KNeighborsClassifier(n_neighbors=3, metric='manhattan') # best result
RF = RandomForestClassifier( bootstrap=False, max_depth=20, max_features='sqrt', min_samples_leaf=1,min_samples_split=2, n_estimators=200) # best result
NB = GaussianNB()
KNN_RF_NB =  ClassifierWrapper(VotingClassifier(
    estimators=[('knn', KNN), ('rf', RF), ('nb', NB)],
    voting='soft'))
```

```python
params  = KNN_RF_NB.calc_params()
params
```
