```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from classifier_wrapper import ClassifierWrapper


KNN = KNeighborsClassifier(n_neighbors=3, metric='manhattan') # best result
RF = RandomForestClassifier( bootstrap=False, max_depth=20, max_features='sqrt', min_samples_leaf=1,min_samples_split=2, n_estimators=200) # best result
DT = DecisionTreeClassifier()
KNN_RF_DT =  ClassifierWrapper(VotingClassifier(
    estimators=[('knn', KNN), ('rf', RF), ('dt', DT)],
    voting='soft'))
```

```python
params  = KNN_RF_DT.calc_params()
params
```
