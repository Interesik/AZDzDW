```python
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# Get data

from sklearn.calibration import LabelEncoder


data = pd.read_csv("metaverse_transactions_dataset.csv").iloc[:, 1:-2]
label_encoder = LabelEncoder()
data = label_encoder.fit_transform(data)
data.head()
```

```python
eps = 3
min_samples = 5

dbscan = DBSCAN(eps=eps, min_samples=min_samples)


# Fit the model to the data
dbscan.fit(data)

# Get the labels of the data points
labels = dbscan.labels_

# Identify the outliers
outliers = np.where(labels == -1)[0]

# Print the number of outliers
print("Number of outliers:", len(outliers))

# Plot the data with the outliers highlighted
plt.scatter(data[:, 0], data[:, 1], c=labels)
plt.scatter(data[outliers, 0], data[outliers, 1], c="red", marker="x")
plt.show()
```
